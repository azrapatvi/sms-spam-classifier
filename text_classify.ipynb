````python
import pandas as pd

messages=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/Complete-Data-Science-With-Machine-Learning-And-NLP-2024/refs/heads/main/26-CompleteNLP%20For%20Machine%20Learning/Practicals/SpamClassifier-master/smsspamcollection/SMSSpamCollection',sep='\t')

messages
````

````python
from nltk.stem import WordNetLemmatizer

lemmatizer=WordNetLemmatizer()
````

````python
import re

from nltk.corpus import stopwords
````

````python
messages.columns=['label','message']
````

````python
corpus=[]
for i in range(len(messages)):
    review=re.sub(r"[^a-zA-Z]"," ",messages['message'][i]) #keeping only characters in sentences
    review=review.lower() #lowercasing the words
    review=review.split()
    
    review=[lemmatizer.lemmatize(word) for word in review]
    corpus.append(review)

print(corpus)
````

````python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
words = []
for sent in corpus:
    # If sent is a list, join it first
    if isinstance(sent, list):
        sent = ' '.join(sent)
    sent_token = sent_tokenize(sent)
    for s in sent_token:
        words.append(simple_preprocess(s))
````

````python
import numpy as np
````

````python
def avg_wrod2vec(doc):
    vectors = [word2vec.wv[word] for word in doc if word in word2vec.wv.index_to_key]
    if len(vectors) == 0:
        return np.zeros(word2vec.vector_size)  # empty message handle
    return np.mean(vectors, axis=0)
````

````python

def preprocess_input(text):
    # Same preprocessing jo tune corpus pe kiya tha
    text = re.sub('[^a-zA-Z]', ' ', text)
    text = text.lower().split()
    text = [lemmatizer.lemmatize(word) for word in text if word not in stopwords.words('english')]
    return text

def predict_message(text):
    # Step 1: Preprocess
    tokens = preprocess_input(text)
    
    # Step 2: Word2Vec avg vector banana
    vector = avg_wrod2vec(tokens).reshape(1, -1)
    
    # Step 3: Predict
    result = random_forest.predict(vector)[0]
    
    return "SPAM 🚨" if result == 1 else "HAM ✅"


new_message = input("enter message:")
print(predict_message(new_message))
````
